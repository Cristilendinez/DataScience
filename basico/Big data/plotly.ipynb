```python
from plotnine.data import mtcars
```

```python
mtcars.head(n=10)
```

```python
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm'))
```

```python
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))
```

```python
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear')
 + theme_xkcd())
```

```python
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear')
 + theme_tufte())
```

```python
(ggplot(mtcars, aes('hp', fill = 'mpg', colour = 'gear')) +
  geom_density(alpha = 0.1) +
  xlim(55, 70))
```
